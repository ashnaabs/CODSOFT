```python
from scipy.stats import zscore
df['age_zscore'] = zscore(df['age'])
df['annualsalary_zscore']=zscore(df['annual Salary'])
df['creditcarddebt_zscore']=zscore(df['credit card debt'])
df['networth_zscore']=zscore(df['net worth'])
df['carpurchaseamount_zscore']=zscore(df['car purchase amount'])

df = df[(df.age_zscore < 3) & (df.age_zscore > -3)]
df = df[(df.annualsalary_zscore < 3) & (df.annualsalary_zscore > -3)]
df = df[(df.creditcarddebt_zscore < 3) & (df.creditcarddebt_zscore > -3)]
df = df[(df.networth_zscore < 3) & (df.networth_zscore > -3)]
df = df[(df.carpurchaseamount_zscore < 3) & (df.carpurchaseamount_zscore > -3)]
```

```python
for i in columns:
  sns.boxplot(df[i])
  plt.show()
```

## Splitting of the data

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
y=y.values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
y=scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42,shuffle=True)
```

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error
```

## Model creation and evaluation

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
knn=KNeighborsRegressor()
svc=SVR()
random=RandomForestRegressor()
desicion=DecisionTreeRegressor()
gbr=GradientBoostingRegressor()
models=[knn,svc,random,desicion,gbr]
```

```python
for i in models:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  mape=mean_absolute_percentage_error(y_test,y_pred)
  r2score=r2_score(y_test,y_pred)
  print(f'Model: {i}')
  print(f'Mean Squared Error: {mape}')
  print(f'r2 score: {r2score}')
  print('-' * 30)
```

```python
model=['knn','svc','random','desicion','gbr']
accuracy=[.9515,.9367,.9990,.9991,.9993]
plt.plot(model,accuracy, marker='o', linestyle='-', color='b')
plt.title('Model Accuracy Comparison')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
